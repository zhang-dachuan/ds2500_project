# Inverse distance weighting

```python
# Create some points, with a distance, and a temperature

import pandas as pd

# data_dict = {'place':['A', 'B', 'C', 'D', 'E'],
#              'dist' :[2, 15, 20, 1000, 1500],
#              'temp' :[30.0, 32.0, 38.0, 68.0, 79.0]}

data_dict = {'place':['A', 'B', 'C', 'D', 'E'] * 2,
             'dist' :[2, 15, 20, 1000, 1500] * 2,
             'temp' :[30.0, 32.0, 38.0, 68.0, 79.0] + [34.0, 41.0, 50.0, 68.0, 82.0],
             'day'  :[1]*5 + [2]*5}



data = pd.DataFrame(data_dict)
data
```

```python
p = 1
data['inv_dist'] = 1 / (data.dist ** p)
data
```

```python
data['weighted'] = data.temp * data.inv_dist
data
```

```python
data.groupby('day').weighted.sum() / data.groupby('day').inv_dist.sum()
```

```python
grouped = data.groupby('day').sum()[['inv_dist', 'weighted']]
grouped
```

```python
grouped['final'] = grouped.weighted / grouped.inv_dist
grouped
```
